## Part 2: Attack Simulation

### Building a Safe Attack Scenario Generator

````python
class AttackPhase(Enum):
    """Kill chain phases."""
    INITIAL_ACCESS = "initial_access"
    EXECUTION = "execution"
    PERSISTENCE = "persistence"
    PRIVILEGE_ESCALATION = "privilege_escalation"
    DEFENSE_EVASION = "defense_evasion"
    CREDENTIAL_ACCESS = "credential_access"
    LATERAL_MOVEMENT = "lateral_movement"
    EXFILTRATION = "exfiltration"


@dataclass
class AttackTechnique:
    """A single ATT&CK technique for simulation."""
    technique_id: str
    name: str
    phase: AttackPhase
    description: str
    expected_artifacts: List[str]
    detection_opportunities: List[str]


@dataclass
class AttackScenario:
    """A complete attack scenario for purple team exercise."""
    name: str
    description: str
    threat_actor: str
    objective: str
    techniques: List[AttackTechnique]
    expected_detections: List[str]


class AttackSimulator:
    """Generate SAFE attack simulations for purple team exercises."""
    
    def __init__(self):
        self.technique_library = self._load_techniques()
    
    def _load_techniques(self) -> Dict[str, AttackTechnique]:
        return {
            "T1059.001": AttackTechnique(
                technique_id="T1059.001",
                name="PowerShell",
                phase=AttackPhase.EXECUTION,
                description="Command execution via PowerShell",
                expected_artifacts=[
                    "Windows Event ID 4688",
                    "Windows Event ID 4104",
                    "Sysmon Event ID 1",
                ],
                detection_opportunities=[
                    "PowerShell with -ExecutionPolicy Bypass",
                    "PowerShell with -NoProfile flag",
                ],
            ),
            "T1003.001": AttackTechnique(
                technique_id="T1003.001",
                name="LSASS Memory",
                phase=AttackPhase.CREDENTIAL_ACCESS,
                description="Credential dumping from LSASS",
                expected_artifacts=[
                    "Process accessing lsass.exe",
                    "Sysmon Event ID 10",
                ],
                detection_opportunities=[
                    "Non-system process accessing LSASS",
                    "Known tools (mimikatz) execution",
                ],
            ),
            "T1021.002": AttackTechnique(
                technique_id="T1021.002",
                name="SMB/Windows Admin Shares",
                phase=AttackPhase.LATERAL_MOVEMENT,
                description="Lateral movement via SMB",
                expected_artifacts=[
                    "Windows Event ID 5140",
                    "Windows Event ID 4624 Type 3",
                ],
                detection_opportunities=[
                    "Access to C$, ADMIN$ shares",
                    "Unusual SMB traffic patterns",
                ],
            ),
        }
    
    def generate_scenario(self, threat_actor: str, objective: str, 
                          techniques: List[str]) -> AttackScenario:
        """Generate a complete attack scenario."""
        scenario_techniques = []
        expected_detections = []
        
        for tech_id in techniques:
            if tech_id in self.technique_library:
                tech = self.technique_library[tech_id]
                scenario_techniques.append(tech)
                expected_detections.extend(tech.detection_opportunities)
        
        return AttackScenario(
            name=f"{threat_actor} Simulation",
            description=f"Purple team exercise simulating {threat_actor}",
            threat_actor=threat_actor,
            objective=objective,
            techniques=scenario_techniques,
            expected_detections=list(set(expected_detections)),
        )

# Example
simulator = AttackSimulator()
scenario = simulator.generate_scenario(
    threat_actor="APT29",
    objective="Credential theft and lateral movement",
    techniques=["T1059.001", "T1003.001", "T1021.002"]
)

print(f"Scenario: {scenario.name}")
print(f"Objective: {scenario.objective}")
print(f"\nTechniques ({len(scenario.techniques)}):")
for tech in scenario.techniques:
    print(f"  â€¢ {tech.technique_id}: {tech.name}")
````

## Part 3: Detection Gap Analysis

````python
@dataclass
class DetectionRule:
    """A detection rule with ATT&CK mapping."""
    rule_id: str
    name: str
    techniques: List[str]
    enabled: bool = True


@dataclass
class GapAnalysis:
    """Results of detection gap analysis."""
    total_techniques: int
    covered_techniques: int
    coverage_percentage: float
    critical_gaps: List[str]
    recommendations: List[str]


class DetectionGapAnalyzer:
    """Analyze detection coverage against MITRE ATT&CK."""
    
    def __init__(self):
        self.attack_matrix = {
            "execution": {
                "T1059.001": {"name": "PowerShell", "priority": "critical"},
                "T1059.003": {"name": "Windows Command Shell", "priority": "high"},
            },
            "credential_access": {
                "T1003.001": {"name": "LSASS Memory", "priority": "critical"},
                "T1558.003": {"name": "Kerberoasting", "priority": "high"},
            },
            "lateral_movement": {
                "T1021.002": {"name": "SMB Admin Shares", "priority": "high"},
                "T1550.002": {"name": "Pass the Hash", "priority": "critical"},
            },
        }
    
    def analyze_coverage(self, rules: List[DetectionRule]) -> GapAnalysis:
        """Analyze detection coverage."""
        covered = set()
        for rule in rules:
            if rule.enabled:
                covered.update(rule.techniques)
        
        total = 0
        gaps = []
        
        for tactic, techniques in self.attack_matrix.items():
            total += len(techniques)
            for tech_id, info in techniques.items():
                if tech_id not in covered and info["priority"] in ["critical", "high"]:
                    gaps.append(f"{tech_id}: {info['name']}")
        
        coverage = len(covered) / total * 100 if total else 0
        
        return GapAnalysis(
            total_techniques=total,
            covered_techniques=len(covered),
            coverage_percentage=coverage,
            critical_gaps=gaps,
            recommendations=[f"Address {len(gaps)} high-priority gaps"] if gaps else [],
        )

# Example
analyzer = DetectionGapAnalyzer()

sample_rules = [
    DetectionRule("R001", "PowerShell Detection", ["T1059.001"]),
    DetectionRule("R002", "LSASS Access", ["T1003.001"]),
]

analysis = analyzer.analyze_coverage(sample_rules)
print(f"Coverage: {analysis.coverage_percentage:.1f}%")
print(f"\nCritical Gaps:")
for gap in analysis.critical_gaps:
    print(f"  ðŸ”´ {gap}")
````

## Part 4: Purple Team Report Generator

````python
class FindingSeverity(Enum):
    CRITICAL = "critical"
    HIGH = "high"
    MEDIUM = "medium"
    LOW = "low"


@dataclass
class PurpleTeamFinding:
    """A finding from a purple team exercise."""
    id: str
    title: str
    severity: FindingSeverity
    technique_id: str
    detection_result: str  # detected, not_detected, partially_detected
    recommendations: List[str]


class PurpleTeamReporter:
    """Generate purple team reports."""
    
    def generate_report(self, exercise_name: str, findings: List[PurpleTeamFinding]) -> dict:
        """Generate a comprehensive report."""
        total = len(findings)
        detected = sum(1 for f in findings if f.detection_result == "detected")
        detection_rate = detected / total * 100 if total else 0
        
        critical_gaps = [
            f.title for f in findings
            if f.detection_result == "not_detected"
            and f.severity in [FindingSeverity.CRITICAL, FindingSeverity.HIGH]
        ]
        
        score = self._calculate_score(findings)
        
        return {
            "title": f"Purple Team Report: {exercise_name}",
            "date": datetime.now().strftime("%Y-%m-%d"),
            "detection_rate": detection_rate,
            "overall_score": score,
            "critical_gaps": critical_gaps,
            "assessment": self._get_assessment(score),
        }
    
    def _calculate_score(self, findings: List[PurpleTeamFinding]) -> float:
        if not findings:
            return 0.0
        
        weights = {
            FindingSeverity.CRITICAL: 5,
            FindingSeverity.HIGH: 4,
            FindingSeverity.MEDIUM: 2,
            FindingSeverity.LOW: 1,
        }
        
        total_weight = sum(weights[f.severity] for f in findings)
        detected_weight = sum(
            weights[f.severity] for f in findings
            if f.detection_result == "detected"
        )
        
        return (detected_weight / total_weight * 100) if total_weight else 0
    
    def _get_assessment(self, score: float) -> str:
        if score >= 80:
            return "STRONG - Minimal critical gaps"
        elif score >= 60:
            return "MODERATE - Several improvements needed"
        return "NEEDS IMPROVEMENT - Significant gaps"

# Example
reporter = PurpleTeamReporter()

sample_findings = [
    PurpleTeamFinding("F001", "PowerShell Bypass", FindingSeverity.HIGH, 
                      "T1059.001", "detected", []),
    PurpleTeamFinding("F002", "LSASS Access", FindingSeverity.CRITICAL, 
                      "T1003.001", "not_detected", ["Add Sysmon monitoring"]),
    PurpleTeamFinding("F003", "SMB Lateral Movement", FindingSeverity.HIGH, 
                      "T1021.002", "partially_detected", []),
]

report = reporter.generate_report("Q1 2026 Exercise", sample_findings)
print(f"Report: {report['title']}")
print(f"Detection Rate: {report['detection_rate']:.0f}%")
print(f"Overall Score: {report['overall_score']:.0f}/100")
print(f"Assessment: {report['assessment']}")
````
